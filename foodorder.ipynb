```python
import random

# Let's say you have a list of items
entrance = ['I want to order', 'send me', 'I would like', 'can we get']
adress = ['my home','my work place','my home two']
payy = ['cash', 'with credit card', 'with debit','with card']
df = pd.DataFrame()
df['text']= []
```

```python
df = pd.DataFrame()
sentences=[]
```

```python
def foodordergenerator():
    # To pick an item randomly, you can use the choice() function
    entr = random.choice(entrance)
    adres= random.choice(adress)
    pay= random.choice(payy)

    number= dfnumber.sample().text.values[0]
    size= dfsize.sample().text.values[0]
    extras= dfextras.sample().text.values[0]
    drink= dfdrink.sample().text.values[0]
    food= dffood.sample().text.values[0]
    try: 
        # Combine the values from the rows into a single string
         sentence = ' '.join([entr, number, food, 'with', extras, 'and', number, size, drink, 'send it to', adres, 'I will pay with', pay])
    except TypeError:
         sentence =''
    return sentence
```

```python
foodordergenerator()
```

```python
for _ in range (1000):
      a = foodordergenerator()
      sentences.append(a)
```

```python
# Loop through the sentences
for sentence in sentences:
    # Add the sentence as a row to the DataFrame, and assign a value to the 'label' column
    df = df.append({'text': sentence},ignore_index=True)
```

```python
def ner(text):
    model = NERModel("bert", "/content/drive/MyDrive/NLP/foodordereng",use_cuda=False)
    trans_table = text.maketrans('', '', string.punctuation)
    text = text.translate(trans_table)

    text=text.lower()
    
    prediction, model_output = model.predict([text])
    filtered_output = [[{v: k} for d in sublist for k, v in d.items() if v.startswith("B-")] for sublist in prediction]
    entities = []
    for sublist in filtered_output:
        for d in sublist:
            for k, v in d.items():
                label = k.split("-")[1]
                entities.append((label, v))
    return entities
```
